```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Mail_Customers.csv')
```

```python
x = data[['CustomerID', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(x_scaled)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_scaled[:, 0], x_scaled[:, 1], x_scaled[:, 2], c=labels, cmap='viridis')
ax.scatter(centers[:, 0], centers[:, 1], centers[:, 2], c='red', s=200, alpha=0.75)
ax.set_xlabel('CustomerID')
ax.set_ylabel('Age')
ax.set_zlabel('Spending Score')
plt.title('Annual Income')
plt.show()
```
